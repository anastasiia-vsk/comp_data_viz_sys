3. Замініть значення "—" в таблиці значеннями NaN 

```python
# не зовсім зрозуміла формулювання завдання, бо в мене в таблиці немає значень "–", вони одразу конвертувалися в NaN
# це видно візуально, та якби був великий датафрейм, могла б перевірити аналітично так:
rows_with_dash = df[df.apply(lambda row: row.astype(str).str.contains('—').any(), axis=1)]
rows_with_dash
```

```python
# Якби значення "–" були, я б замінила їх так:
df = df.replace("—", np.nan)
df
```

```python
# тому я подивилася на пусті значення іншим шляхом
rows_with_nan = df[df.isna().any(axis=1)]
rows_with_nan
```

```python
# також, їх можна було б виявити таким чином:
df.isna().sum()
```

8. Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком  (метод fillna) 

```python
df_filled = df.fillna(df[['1950', '1960', '1970', '2014']].mean())
df_filled
```

5. Замініть типи нечислових стовпців числовими стовпцями. Це стовпці, у яких було розташовано символ "—" 

```python
# тут можна було б змінювати кожну колонку окремо
# df_filled['1950'] = df_filled['1950'].astype(int)
# df_filled['1960'] = df_filled['1960'].astype(int)
# і так далі...

# але я люблю все оптимізовувати
columns_to_convert = ['1950', '1960', '1970', '2014']

for column in columns_to_convert:
    df_filled[column] = df_filled[column].astype(int)
    
display(df_filled.tail(5))
display(df_filled.dtypes)
```

9. Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні 

```python
# переконалася, що беру правильний датафрейм
display(df_filled.head(), df_filled.tail())
```

```python
country_Ukraine = df_filled[df_filled['регіон'] == 'Україна']
mean_Ukraine = country_Ukraine['2014'].mean()
print("Середнє значення по Україні:", mean_Ukraine)
```
