1. Побудуйте звичайну лінійну діаграму на основі  pd.DataFrame з двома стовпцями за допомогою бібліотеки Altair для функції `y=xcosx` для значень x в діапазоні від 0 до 99. Додайте відповідні мітки для вісей x та y на діаграмі. Підпишіть вісі і назву графіку. 

```python
x_values = np.arange(0, 100)
y_values = x_values * np.cos(x_values)
df = pd.DataFrame({"x": x_values, "y": y_values})
```

```python
line_chart = (
    alt.Chart(df)
    .mark_line()
    .encode(x="x:Q", y="y:Q")
    .properties(title="Графік функції y = xcos(x)", width=800, height=300)
    .configure_axis(labelFontSize=12, titleFontSize=14)
)

line_chart
```

2. Створіть горизонтальну діаграму з мітками на основі даних про врожайність пшениці (data.wheat()) за допомогою бібліотеки  Altair. Задайте вісь x для врожайності пшениці, вісь y для років, а текстові мітки відображайте значення врожайності пшениці на відповідних роках.  

```python
source = data.wheat()
source.head()
```

```python
bar_chart = (
    alt.Chart(source)
    .mark_bar()
    .encode(x="wheat:Q", y="year:O", text="wheat:Q", color=alt.value("darkblue"))
    .properties(title="Врожайність пшениці за роками", width=600, height=700)
)

chart = bar_chart.mark_bar() + bar_chart.mark_text(align="left", dx=2)

chart = chart.configure_axis(labelFontSize=14, titleFontSize=14).configure_title(
    fontSize=20
)

chart = chart.encode(
    x=alt.X("wheat:Q", title="Врожайність пшениці"), y=alt.Y("year:O", title="Роки")
)

chart.display()
```

3. Побудуйте гістограму із від'ємними значеннями на основі даних про зміни в зайнятості в США (data.us_employment()) за допомогою бібліотеки Altair. Вісь x відповідає місяцям, вісь y відображає зміни в зайнятості (nonfarm_change). Використовуйте два різних кольори для відображення від'ємних та додатних значень змін в зайнятості. Встановіть ширину діаграми на 500 пікселів. 

```python
employment_data = data.us_employment()
employment_data.head()
```

```python
histogram = (
    alt.Chart(employment_data)
    .mark_bar()
    .encode(
        x=alt.X("month:T", title="Місяць"),
        y=alt.Y("nonfarm_change:Q", title="Зміни в зайнятості"),
        color=alt.condition(
            alt.datum.nonfarm_change > 0,
            alt.value("steelblue"),
            alt.value("orange"),
        ),
    )
    .properties(width=500, title="Зміни в зайнятості в США")
)

histogram
```

4. Побудуйте гістограму з групуванням на основі даних про врожай ячменю (data.barley()). Вісь x представляє роки, вісь y відображає сумарний врожай (yield), кольорова шкала вказує на різні роки, а групування (стовпці) відбувається за різними місцями (site). 

```python
barley_data = data.barley()
barley_data.head()
```

```python
histogram_barley = (
    alt.Chart(barley_data)
    .mark_bar()
    .encode(x="year:N", y="sum(yield):Q", color="year:N", column="site:N")
    .properties(
        width=100, 
        title="Сумарний врожай ячменю за рік та місце",
    )
)

histogram_barley
```

5. Створіть багатошарову діаграму на основі даних про виробництво електроенергії в штаті Айова (data.iowa_electricity()). Вісь x представляє роки, вісь y відображає чисте виробництво електроенергії (net_generation), а кожен шар діаграми позначає різні джерела енергії (source). Встановіть прозорість (opacity) для стовпців на рівні 0.7. 

```python
source = data.iowa_electricity()
source.head()
```

```python
source["year"] = pd.to_datetime(source["year"]).dt.year
source.head()
```

```python
chart_iowa = (
    alt.Chart(source)
    .mark_bar(opacity=0.7)
    .encode(
        x=alt.X("year:O", title="Роки", axis=alt.Axis(labelAngle=-45)),
        y=alt.Y("net_generation:Q", title="Кількість електроенергії"),
        color="source:N",
    )
    .properties(title="Виробництво електроенергії в штаті Айова", width=700, height=400)
    .configure_axis(labelFontSize=12, titleFontSize=14)
    .configure_title(fontSize=20)
)

chart_iowa.display()
```

6. Побудуйте лінійний графік з маркерами на основі даних про акції (data.stocks()) . Вісь x представляє дати, вісь y відображає ціни акцій (price), а кожен графік позначається символом (symbol). Встановіть маркери як неповні, заповнюючи їх білим кольором, та використовуйте різні кольори для різних символів акцій.

7. Створіть діаграму розсіювання для двовимірної гістограми на основі даних про фільми (data.movies.url). Вісь X відображає рейтинг фільмів на IMDB (IMDB_Rating), вісь Y відображає рейтинг фільмів на Rotten Tomatoes (Rotten_Tomatoes_Rating).  

```python
movies_data = data.movies.url
```

```python
scatterplot = (
    alt.Chart(movies_data)
    .mark_point()
    .encode(
        x=alt.X("IMDB_Rating:Q", bin=alt.Bin(maxbins=30), title="IMDB Rating"),
        y=alt.Y(
            "Rotten_Tomatoes_Rating:Q",
            bin=alt.Bin(maxbins=30),
            title="Rotten Tomatoes Rating",
        ),
        color=alt.Color("count()", scale=alt.Scale(scheme="oranges"), title="Count"),
    )
    .properties(width=600, height=400, title="Двовимірна гістограма рейтингів фільмів")
)

scatterplot
```

```python
scatter_plot = (
    alt.Chart(df_education)
    .mark_circle()
    .encode(
        x="gross_primary_education_enrollment:Q",
        y="gross_tertiary_education_enrollment:Q",
        tooltip=[
            "countries_and_areas:N",
            "gross_primary_education_enrollment:Q",
            "gross_tertiary_education_enrollment:Q",
        ],
    )
    .properties(
        title="Графік розсіювання між Gross Primary та Gross Tertiary Education Enrollment",
        width=700,
        height=400,
    )
    .configure_axis(labelFontSize=12, titleFontSize=14)
    .configure_title(fontSize=16)
)

scatter_plot.display()
```

- Порівняти рейтинги грамотності чоловіків та жінок ('Youth_15_24_Literacy_Rate_Male' та 'Youth_15_24_Literacy_Rate_Female') за допомогою горизонтальної стовпчастої діаграми. 

```python
# Чоловіки
chart_male = (
    alt.Chart(df_education)
    .mark_bar()
    .encode(
        x=alt.X("youth_15_24_literacy_rate_male:Q", title="Грамотність чоловіків"),
        y=alt.Y("countries_and_areas:N", title="Країни та регіони"),
        color=alt.value("skyblue"), 
        tooltip=["youth_15_24_literacy_rate_male:Q"],
    )
    .properties(
        title="Грамотність чоловіків у країнах та регіонах", width=400, height=2000
    )
)


# Жінки
chart_female = (
    alt.Chart(df_education)
    .mark_bar()
    .encode(
        x=alt.X("youth_15_24_literacy_rate_female:Q", title="Грамотність жінок"),
        y=alt.Y("countries_and_areas:N", title="Країни та регіони"),
        color=alt.value("salmon"),
        tooltip=[
            "youth_15_24_literacy_rate_female:Q"
        ], 
    )
    .properties(title="Грамотність жінок у країнах та регіонах", width=400, height=2000)
)

chart = (
    alt.HConcatChart(hconcat=[chart_male, chart_female])
    .configure_axis(labelFontSize=10, titleFontSize=14)
    .configure_title(fontSize=16)
)

chart.display()
```
